```python
import os

import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib.ticker import FormatStrFormatter
```

```python
sample_dict_Ti_only = {
    "0p0": "ytso_0_try6_retry12",
    "0p2": "ytso_0p2_try8_8_1_2_retry_6_optimal",
    "0p5": "ytso_0p5_try2_scale4_try23_LP_2",
    "0p8": "ytso_0p8_try2_scale4_try3_optimal"
}

sample_dict_Sn_only = {
    "0p2": "ytso_0p2_try8_8_1_2_retry_6_optimal",
    "0p5": "ytso_0p5_try2_scale4_try23_LP_2",
    "0p8": "ytso_0p8_try2_scale4_try3_optimal",
    "1p0": "ytso_1p0_try2_scale2_FSOnly_Optimal"
}

sample_dict_all = {
    "0p0": "ytso_0_try6_retry12",
    "0p2": "ytso_0p2_try8_8_1_2_retry_6_optimal",
    "0p5": "ytso_0p5_try2_scale4_try23_LP_2",
    "0p8": "ytso_0p8_try2_scale4_try3_optimal",
    "1p0": "ytso_1p0_try2_scale2_FSOnly_Optimal"
}

line_style = {
    "y1": 'o-',
    "y2": '^-',
    "y3": 's-',
    "y4": 'X-'
}

color_dict = {
    "y1": 'red',
    "y2": 'peru',
    "y3": 'orange',
    "y4": 'olivedrab'
}
```

```python
f = plt.figure(figsize=(18, 15))
gs = gridspec.GridSpec(3, 1, height_ratios=[1 for _ in range(3)], hspace=0)
axes = list()
for i in range(3):
    axes.append(plt.subplot(gs[i, 0]))

# Ti only
x = [0.0, 0.2, 0.5, 0.8]
y1 = list()
y2 = list()
y3 = list()
y4 = list()
for key, item in sample_dict_Ti_only.items():
    with open(os.path.join(item, "ani_Ti.out"), "r") as ff:
        lines = ff.readlines()
    y1.append(float(lines[0].strip().split()[0]))
    y2.append(float(lines[1].strip().split()[0]))
    y3.append(float(lines[2].strip().split()[0]))
    y4.append(float(lines[3].strip().split()[0]))

axes[0].plot(
    x,
    y1,
    line_style["y1"],
    color=color_dict["y1"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[0].plot(
    x,
    y2,
    line_style["y2"],
    color=color_dict["y2"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[0].plot(
    x,
    y3,
    line_style["y3"],
    color=color_dict["y3"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[0].plot(
    x,
    y4,
    line_style["y4"],
    color=color_dict["y4"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)

# Sn only
x = [0.2, 0.5, 0.8, 1.0]
y1 = list()
y2 = list()
y3 = list()
y4 = list()
for key, item in sample_dict_Sn_only.items():
    with open(os.path.join(item, "ani_Sn.out"), "r") as ff:
        lines = ff.readlines()
    y1.append(float(lines[0].strip().split()[0]))
    y2.append(float(lines[1].strip().split()[0]))
    y3.append(float(lines[2].strip().split()[0]))
    y4.append(float(lines[3].strip().split()[0]))

axes[1].plot(
    x,
    y1,
    line_style["y1"],
    color=color_dict["y1"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[1].plot(
    x,
    y2,
    line_style["y2"],
    color=color_dict["y2"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[1].plot(
    x,
    y3,
    line_style["y3"],
    color=color_dict["y3"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[1].plot(
    x,
    y4,
    line_style["y4"],
    color=color_dict["y4"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)

# Ti & Sn
x = [0.0, 0.2, 0.5, 0.8, 1.0]
y1 = list()
y2 = list()
y3 = list()
y4 = list()
for key, item in sample_dict_all.items():
    with open(os.path.join(item, "ani_all.out"), "r") as ff:
        lines = ff.readlines()
    y1.append(float(lines[0].strip().split()[0]))
    y2.append(float(lines[1].strip().split()[0]))
    y3.append(float(lines[2].strip().split()[0]))
    y4.append(float(lines[3].strip().split()[0]))

axes[2].plot(
    x,
    y1,
    line_style["y1"],
    color=color_dict["y1"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[2].plot(
    x,
    y2,
    line_style["y2"],
    color=color_dict["y2"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[2].plot(
    x,
    y3,
    line_style["y3"],
    color=color_dict["y3"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[2].plot(
    x,
    y4,
    line_style["y4"],
    color=color_dict["y4"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
```
