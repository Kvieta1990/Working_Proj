```python
import os

import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
sample_dict_Ti_only = {
    "0p0": "ytso_0_try6_retry12",
    "0p2": "ytso_0p2_try8_8_1_2_retry_6_optimal",
    "0p5": "ytso_0p5_try2_scale4_try23_LP_2",
    "0p8": "ytso_0p8_try2_scale4_try3_optimal"
}

sample_dict_Sn_only = {
    "0p2": "ytso_0p2_try8_8_1_2_retry_6_optimal",
    "0p5": "ytso_0p5_try2_scale4_try23_LP_2",
    "0p8": "ytso_0p8_try2_scale4_try3_optimal",
    "1p0": "ytso_1p0_try2_scale2_FSOnly_Optimal"
}

sample_dict_all = {
    "0p0": "ytso_0_try6_retry12",
    "0p2": "ytso_0p2_try8_8_1_2_retry_6_optimal",
    "0p5": "ytso_0p5_try2_scale4_try23_LP_2",
    "0p8": "ytso_0p8_try2_scale4_try3_optimal",
    "1p0": "ytso_1p0_try2_scale2_FSOnly_Optimal"
}

line_style = {
    "y1": 'o-',
    "y2": '^-',
    "y3": 's-',
    "y4": 'X-'
}

color_dict = {
    "y1": 'red',
    "y2": 'peru',
    "y3": 'orange',
    "y4": 'olivedrab'
}
```

```python
f = plt.figure(figsize=(18, 15))
gs = gridspec.GridSpec(3, 1, height_ratios=[1 for _ in range(3)], hspace=0)
axes = list()
for i in range(3):
    axes.append(plt.subplot(gs[i, 0]))

# Ti only
x = [0.0, 0.2, 0.5, 0.8]
y1 = list()
y2 = list()
y3 = list()
y4 = list()
sum_u2 = 0.
for key, item in sample_dict_Ti_only.items():
    with open(os.path.join(item, "ani_Ti.out"), "r") as ff:
        lines = ff.readlines()
    y1.append(float(lines[0].strip().split()[0]))
    y2.append(float(lines[1].strip().split()[0]))
    y3.append(float(lines[2].strip().split()[0]))
    y4.append(float(lines[3].strip().split()[0]))
    sum_u2 += float(lines[4].strip().split()[0])
ave_level = sum_u2 / 4.

axes[0].plot(
    x,
    y1,
    line_style["y1"],
    color=color_dict["y1"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[0].plot(
    x,
    y2,
    line_style["y2"],
    color=color_dict["y2"],
    label=r"$u_{23}$",
    linewidth=2,
    markersize=15,
)
axes[0].plot(
    x,
    y3,
    line_style["y3"],
    color=color_dict["y3"],
    label=r"$u_{13}$",
    linewidth=2,
    markersize=15,
)
axes[0].plot(
    x,
    y4,
    line_style["y4"],
    color=color_dict["y4"],
    label=r"$u_{12}$",
    linewidth=2,
    markersize=15,
)
axes[0].plot(
    [-0.05, 1.05],
    [ave_level, ave_level],
    "k--",
    label=r"$\langle u^2\rangle$",
    linewidth=2,
)

axes[0].set_xlim([-0.05, 1.05])

# Sn only
x = [0.2, 0.5, 0.8, 1.0]
y1 = list()
y2 = list()
y3 = list()
y4 = list()
for key, item in sample_dict_Sn_only.items():
    with open(os.path.join(item, "ani_Sn.out"), "r") as ff:
        lines = ff.readlines()
    y1.append(float(lines[0].strip().split()[0]))
    y2.append(float(lines[1].strip().split()[0]))
    y3.append(float(lines[2].strip().split()[0]))
    y4.append(float(lines[3].strip().split()[0]))

axes[1].plot(
    x,
    y1,
    line_style["y1"],
    color=color_dict["y1"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[1].plot(
    x,
    y2,
    line_style["y2"],
    color=color_dict["y2"],
    label=r"$u_{23}$",
    linewidth=2,
    markersize=15,
)
axes[1].plot(
    x,
    y3,
    line_style["y3"],
    color=color_dict["y3"],
    label=r"$u_{13}$",
    linewidth=2,
    markersize=15,
)
axes[1].plot(
    x,
    y4,
    line_style["y4"],
    color=color_dict["y4"],
    label=r"$u_{12}$",
    linewidth=2,
    markersize=15,
)
axes[1].plot(
    [-0.05, 1.05],
    [ave_level, ave_level],
    "k--",
    label=r"$\langle u^2\rangle$",
    linewidth=2,
)

axes[1].set_xlim([-0.05, 1.05])

# Ti & Sn
x = [0.0, 0.2, 0.5, 0.8, 1.0]
y1 = list()
y2 = list()
y3 = list()
y4 = list()
for key, item in sample_dict_all.items():
    with open(os.path.join(item, "ani_all.out"), "r") as ff:
        lines = ff.readlines()
    y1.append(float(lines[0].strip().split()[0]))
    y2.append(float(lines[1].strip().split()[0]))
    y3.append(float(lines[2].strip().split()[0]))
    y4.append(float(lines[3].strip().split()[0]))

axes[2].plot(
    x,
    y1,
    line_style["y1"],
    color=color_dict["y1"],
    label=r"$\sigma_{\langle u^2\rangle}$",
    linewidth=2,
    markersize=15,
)
axes[2].plot(
    x,
    y2,
    line_style["y2"],
    color=color_dict["y2"],
    label=r"$u_{23}$",
    linewidth=2,
    markersize=15,
)
axes[2].plot(
    x,
    y3,
    line_style["y3"],
    color=color_dict["y3"],
    label=r"$u_{13}$",
    linewidth=2,
    markersize=15,
)
axes[2].plot(
    x,
    y4,
    line_style["y4"],
    color=color_dict["y4"],
    label=r"$u_{12}$",
    linewidth=2,
    markersize=15,
)
axes[2].plot(
    [-0.05, 1.05],
    [ave_level, ave_level],
    "k--",
    label=r"$\langle u^2\rangle$",
    linewidth=2,
)

axes[2].set_xlim([-0.05, 1.05])

axes[0].tick_params(axis="x", direction="out", length=12, width=1, labelsize=30)
axes[1].tick_params(axis="x", direction="out", length=12, width=1, labelsize=30)
axes[2].tick_params(axis="x", direction="out", length=12, width=1, labelsize=30)
axes[0].set_xticklabels([])
axes[1].set_xticklabels([])
axes[1].xaxis.set_ticks_position("both")
axes[2].xaxis.set_ticks_position("both")

for i in range(3):
    axes[i].tick_params(axis="y", direction="out", length=12, width=1, labelsize=30)
    axes[i].spines["left"].set_linewidth(1)
    axes[i].spines["right"].set_linewidth(1)
    axes[i].spines["top"].set_linewidth(1)
    axes[i].spines["bottom"].set_linewidth(1)
    axes[i].legend(fontsize=25, frameon=False)

axes[2].set_xlabel(r"Doping level", fontdict={"size": 30})
axes[0].set_ylabel(r"Atomic Displacement Parameter ($\AA^2$)", fontdict={"size": 30})
axes[0].yaxis.set_label_coords(-0.1, -0.55)
plt.tight_layout()

_ = f.savefig('ADP_all.png', format='png', dpi=600)
```
